```python
%matplotlib inline
import ipywidgets as widgets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (15,7)
import seaborn as sns
import sys
from io import StringIO
from IPython.display import display, Markdown
from model import GUI_substation


def GUI(): 
    #-----------Choose scenario tab----------------------------------
    runbtn = widgets.Button(
            description='Run Simulation',
            disabled=False,
            button_style='success', # 'success', 'info', 'warning', 'danger' or ''
            tooltip='Click me',
            icon='', # (FontAwesome names without the `fa-` prefix)
            layout = {'font-size':'24px',
                     'margin': '20px 0 10px 0'}
            )

    scenario_selection = widgets.ToggleButtons(
             options=['Robust', 'Inclusive', 'Individualistic', 'Modern'],
             description='',
             disabled=False,
             button_style='', # 'success', 'info', 'warning', 'danger' or ''
             tooltips=['Inspired by Forte', 'Inspired by Legato', 'Inspired by Espressivo', 'Inspired by Vivace'],
             #icons=['check'] * 4
             layout={'width': 'max-content',
                    'margin':'15px',
                    'margin-bottom': '30px'},)

    chooselayout = widgets.Layout(display='flex',
                        flex_flow='column',
                        align_items='center',
                        border='none',
                        width='100%',
                        )

    scenariobox = widgets.VBox(
        [widgets.HTML(
        value="<h2> Choose a scenario </h2>",
        ),
         scenario_selection, 
         runbtn,],
        layout = chooselayout
        )

    #------------------ Build scenario tab ----------------------------------

    region_dropdown = widgets.Dropdown(options=['Stockholm'],
                                       description='Region:',
                                       disabled=False,
                                      )
    regionbox = widgets.HBox([region_dropdown], layout = {'margin':'0 0 0 700px'})

    housenew = widgets.IntText(value=0,layout={'width': '70px'})
    housenewbox = widgets.Box([widgets.Label('New houses (electrical heating):'),
                            housenew],)

    houseold = widgets.IntText(value=0,layout={'width': '70px'})
    houseoldbox = widgets.Box([widgets.Label('Old houses (electrical heating):'),
                            houseold],)

    housemix = widgets.IntText(value=0,layout={'width': '70px'})
    housemixbox = widgets.Box([widgets.Label('Mixed age houses (district heating):'),
                            housemix],)

    apart = widgets.IntText(value=0,layout={'width': '70px'})
    apartbox = widgets.Box([widgets.Label('New apartments (district heating):'),
                          apart])

    apartbox = widgets.Box([widgets.Label('New apartments (district heating):'),
                          apart])

    resloadbox = widgets.VBox([widgets.HTML('<h4>Residential buildings</h4>'),
                               widgets.VBox([housenewbox,
                                             houseoldbox,
                                             housemixbox,
                                             apartbox],)
                              ],layout = {'margin':'15px'})
    
    office800 = widgets.IntText(value = 0,layout={'width': '100px'})
    office12402 = widgets.IntText(value=0,layout={'width': '100px'})
    office12667 = widgets.IntText(value=0,layout={'width': '100px'})
    office28246 = widgets.IntText(value=0,layout={'width': '100px'})
    office66799 = widgets.IntText(value=0,layout={'width': '100px'})
    office73620 = widgets.IntText(value=0,layout={'width': '100px'})
    customsize = widgets.IntText(value=0,layout={'width': '67px'})
    officecustom = widgets.IntText(value=0,layout={'width': '100px'})
    officebox = widgets.VBox([widgets.HTML(value='<h4>Offices</h4>'),
                              widgets.HTML(value='<p> Choose number of offices in each size. m2 referes to the heated area (atemp) of the buildings. </p>'),
                              widgets.HBox([
                                  widgets.VBox([widgets.Label('800 m2'),
                                               office800], layout = {'padding':'0 10px 0 0'}),
                                  widgets.VBox([widgets.Label('12402 m2'),
                                               office12402], layout = {'padding':'0 10px 0 10px'}),
                                  widgets.VBox([widgets.Label('12667 m2'),
                                               office12667],layout = {'padding':'0 10px 0 10px'}),
                                  widgets.VBox([widgets.Label('28246 m2'),
                                               office28246],layout = {'padding':'0 10px 0 10px'}),
                                  widgets.VBox([widgets.Label('66799 m2'),
                                               office66799],layout = {'padding':'0 10px 0 10px'}),
                                  widgets.VBox([widgets.Label('73620 m2'),
                                               office73620],layout = {'padding':'0 10px 0 10px'}),
                                  widgets.VBox([widgets.HBox([widgets.Label('m2  ', layout = {'padding':'0 10px 0 0'}), customsize]),
                                               officecustom],layout = {'padding':'0 10px 0 10px'}),
                                  
                              ],)
                             ],
                             layout = {'margin':'15px',}
                            )

    PV69 = widgets.IntText(value=0,
                             min=0, 
                             #description='10.5 kWp',
                             disabled=False,
                             layout={'width': '150px'})
    PV400 = widgets.IntText(value=0,
                             min=0,
                             #description='66 kWp:',
                             disabled=False,
                             layout={'width': '150px'})
    PV868 = widgets.IntText(value=0,
                             min=0,
                             #description='143 kWp:',
                             disabled=False,
                             layout={'width': '150px'})
    pvsize = widgets.IntText(value=0,
                             min=0,
                             disabled=False,
                             layout={'width': '100px'})
    pvcustom = widgets.IntText(value=0,
                             min=0,
                             #description='',
                             disabled=False,
                             layout={'width': '150px'})
    PVbox = widgets.VBox([widgets.HTML('<h4>PV installations</h4>'),
                         widgets.HBox([widgets.VBox([widgets.Label('10.5 kWp'),PV69],
                                                    layout = {'padding':'0 0 0 10px'}),
                                       widgets.VBox([widgets.Label('66 kWp'),PV400],
                                                    layout = {'padding':'0 10px 0 10px'}),
                                       widgets.VBox([widgets.Label('143 kWp'),PV868],
                                                    layout = {'padding':'0 10px 0 10px'}),
                                       widgets.VBox([widgets.HBox([
                                           widgets.Label('kWp  ', layout = {'padding':'0 20px 0 0'}),
                                                                                 pvsize]),pvcustom],
                                                    layout = {'padding':'0 10px 0 10px'})])
                         ],
                        layout = {'margin':'15px'})
    
    up = widgets.FileUpload(description=" Select .csv",
                            accept=".csv", 
                            multiple=False)
    
    upbox = widgets.VBox([widgets.HTML('<h4>Custom loads </h4>'),
                         widgets.HTML('<p>Upload a csv-file to include other loads. For compatability the csv needs to have a DateTimeIndex of hourly frequancy and the comsumption time series of one or more loads should be separte columns.  </p>'),
                         up],
                        layout = {'margin':'15px'})

    EVnumber = widgets.IntText(value = 0, 
                               #description = 'Number of EVs in substation:',
                              layout={'width': '70px'})
    EVparking = widgets.IntText(value = 0, 
                               #description = 'Number of Parking spaces in substation:',
                               layout={'width': '70px'})
    EVkwhkm = widgets.FloatText(value = 0.2, 
                               layout={'width': '70px'})

    EVbox = widgets.VBox([widgets.HTML('<h4>Electrical Vehicles </h4>'),
                          widgets.Box([widgets.Label('Number of EVs:'),EVnumber,
                             widgets.Label('Number of parking spaces:'),EVparking,
                             widgets.Label('Mean consumption of EVs (kWh/km):'),EVkwhkm])
                         ],
                       layout = {'margin':'15px'})

    flexcheck = widgets.Checkbox(value=False,
                                 #description='Demand Side Flexibility',
                                 disabled=False,
                                 indent=False
                                )

    flexpercent = widgets.IntText(value=30,
                             min=0,
                             disabled= True,
                            layout={'width': '70px'}
                            )

    flexdays = widgets.IntText(value=17,
                             min=0,
                             disabled= True,
                            layout={'width': '70px'}
                            )
    flexloadpercent = widgets.IntText(value=10,
                             min=0,
                             disabled= True,
                            layout={'width': '70px'}
                            )

    flexbox = widgets.VBox([widgets.HBox([widgets.HTML('<h4>Static Flexibilty</h4>'),flexcheck]),
                         widgets.HBox([widgets.Label('Flex extent (%)'),flexpercent, 
                         widgets.Label('Days to be flexible/year:'),flexdays, 
                         widgets.Label('Percentage of loads that are flexible:'),flexloadpercent,])
                        ], 
                        layout = {'margin':'15px'})
    
    optimalcheck = widgets.Checkbox(value=False,
                                 disabled=False,
                                 indent=False
                                )
    kW = widgets.FloatText(value = 0.0, 
                           disabled = True,
                           layout={'width': '70px'})
    kWh = widgets.FloatText(value = 0.0, 
                            disabled = True,
                            layout={'width': '70px'})
    kWLabel = widgets.Label('maxkW ')
    kWhLabel = widgets.Label('maxkWh ')
    
    flextoggle = widgets.ToggleButtons(
                        options=['Battery', 'Flexible Agents'],
                        disabled=True,
                        button_style='', # 'success', 'info', 'warning', 'danger' or ''
                        tooltips=['Description of Battery', 'Description of Flexible Agents'],
                        layout = {'margin':'0 0 10px 30px'}
                        )
    
    optimalflexbox = widgets.VBox([widgets.HBox([widgets.HTML('<h4>Optimized Flexibilty</h4>'),optimalcheck]),
                        widgets.HBox([widgets.Label('Flexibility Type:'),flextoggle], layout= {}),
                         widgets.VBox([widgets.HBox([kWLabel,kW]),
                                       widgets.HBox([kWhLabel,kWh])])
                        ], 
                        layout = {'margin':'15px'})


    EEcheck = widgets.Checkbox(value=False,
                               description='',
                               disabled=False,
                               indent=False,
                               layout={'font-size': '16px'},
                                )

    EEreduction = widgets.IntSlider(min = 0,
                                    max = 100,
                                    value = 30,
                                    disabled = True, 
                                   )

    EEloads = widgets.IntSlider(min = 0,
                                max = 0,
                                disabled = True,
                               )

    EEbox = widgets.VBox([widgets.HBox([widgets.HTML('<h4>Energy Efficency</h4>'),
                          EEcheck]), 
                             widgets.VBox([widgets.HBox([widgets.Label('Number of loads being energy efficient'),EEloads]),
                                           widgets.HBox([widgets.Label('Reduction of consumption [%]'),EEreduction]),
                                          ],
                        layout={'width': 'max-content'},
                         ), 
                         ],
                       layout = {'margin':'15px'})

    runbtn2 = widgets.Button(
            description='Run Simulation',
            disabled=False,
            button_style='success', # 'success', 'info', 'warning', 'danger' or ''
            tooltip='Click me',
            icon='', # (FontAwesome names without the `fa-` prefix)
            layout = {'font-size':'24px'}
            )


    btnbox = widgets.HBox([runbtn2],
                        layout = {'width': 'max-content',
                                  'align-items':'stretch',
                                  'margin':'15px'})

    custom_layout = widgets.Layout(display='flex',
                                   flex_flow='column',
                                   align_items='stretch',
                                   border='none',
                                   width='99%',
                        )

    create_scenario = widgets.Box([regionbox, 
                                    resloadbox, 
                                    officebox,
                                    upbox,
                                    PVbox,
                                    EVbox, 
                                    EEbox, 
                                    flexbox,
                                    optimalflexbox,
                                    btnbox,
                                   ],
                                  layout = custom_layout)



    #-------------------Events and Output -----------------------------------

    def run_model(event, scenario = None):
        if scenario =='Robust':
            print('Running scenario', scenario,'...')
            region = 'Stockholm'
            # Loads
            resload_dict = {'HouseNew': 10,
                            'HouseOld' : 10, 
                            'HouseDH': 25,
                            'ApartmentNewDH': 55
                           }
            office_dict = {100:1,
                           500:1,
                          }
            # PV
            pv_dict = {0:0,
                      }
            # EV
            num_EV = 6
            num_parkingloc = 6
            mpg_mu = 0.2
            # Energy Efficiency
            is_efficient = True
            efficiency_percent = 0.05 # Reduce values with 5%
            efficient_loads = 90
            # Flex
            is_flex = False
            flex_days = 0
            percent_flex_loads = 0.0
            flex_reduction = 0.0
            flex_only_noDH = True
            # Custom Loads
            custom = pd.DataFrame()
            # Optimal Flex
            optimal_flex = False
            maxkW = 0
            maxkWh = 0
        elif scenario == 'Inclusive':
            print('Running scenario', scenario,'...')
            region = 'Stockholm'
            # Loads
            resload_dict = {'HouseNew': 13,
                            'HouseOld' : 12, 
                            'HouseDH': 20, 
                            'ApartmentNewDH': 55
                           }
            office_dict = {100:1,
                           500:1,
                          }
            # PV
            pv_dict = {69:20,
                       400:19,
                       868:1,
                      }
            # EV
            num_EV = 39
            num_parkingloc = 39
            mpg_mu = 0.2
            # Energy Efficiency
            is_efficient = True
            efficiency_percent = 0.15 # Reduce values with 5%
            efficient_loads = 102
            # Flex
            is_flex = True
            flex_days = 50
            percent_flex_loads = 0.3
            flex_reduction = 0.5
            flex_only_noDH = True
            # Custom Loads
            custom = pd.DataFrame()
            # Optimal Flex
            optimal_flex = True
            maxkW = 40
            maxkWh = 60
        elif scenario == 'Individualistic':
            print('Running scenario', scenario,'...')
            region = 'Stockholm'
            # Loads
            resload_dict = {'HouseNew': 25,
                            'HouseOld' : 25, 
                            'HouseDH': 10, 
                            'ApartmentNewDH': 35
                           }
            office_dict = {100:1,
                           500:1,
                          }
            # PV
            pv_dict = {69:14,
                       400:13,
                       868:1,
                      }
            # EV
            num_EV = 79
            num_parkingloc = 79
            mpg_mu = 0.2
            # Energy Efficiency
            is_efficient = True
            efficiency_percent = 0.10 # Reduce values with 5%
            efficient_loads = 97
            # Flex
            is_flex = True
            flex_days = 50
            percent_flex_loads = 0.3
            flex_reduction = 0.2
            flex_only_noDH = True
            # Custom Loads
            custom = pd.DataFrame()
            # Optimal Flex
            optimal_flex = False
            maxkW = 0
            maxkWh = 0
        elif scenario == 'Modern':
            print('Running scenario', scenario,'...')
            region = 'Stockholm'
            # Loads
            resload_dict = {'HouseNew': 16,
                            'HouseOld' : 16, 
                            'HouseDH': 13, 
                            'ApartmentNewDH': 55
                           }
            office_dict = {100:1,
                           500:1,
                          }
            # PV
            pv_dict = {69:25,
                       400:24,
                       868:1,
                      }
            # EV
            num_EV = 118
            num_parkingloc = 118
            mpg_mu = 0.15
            # Energy Efficiency
            is_efficient = True
            efficiency_percent = 0.15 # Reduce values with 5%
            efficient_loads = 102
            # Flex
            is_flex = False
            flex_days = 0
            percent_flex_loads = 0.0
            flex_reduction = 0.0
            flex_only_noDH = True
            # Custom Loads
            custom = pd.DataFrame()
            # Optimal Flex
            optimal_flex = True
            maxkW = 60
            maxkWh = 80
        else:
            print('Running custom scenario...')
            region = 'Stockholm'
            # Loads
            office_dict = {800:office800.value, 
                           12402:office12402.value, 
                           12667:office12667.value, 
                           28246:office28246.value, 
                           66799:office66799.value,
                           73620:office73620.value,
                          customsize.value:officecustom.value}
            resload_dict = {'HouseNew': housenew.value,
                            'HouseOld': houseold.value,
                            'HouseDH': housemix.value, 
                            'ApartmentNewDH': apart.value,
                           }

            # PV
            pv_dict = {69 : PV69.value,
                       400 : PV400.value,
                       868 : PV868.value,
                    pvsize.value : pvcustom.value}

            # EV
            num_EV = EVnumber.value
            num_parkingloc = EVparking.value

            # EV
            num_EV = EVnumber.value
            num_parkingloc = EVparking.value
            mpg_mu = EVkwhkm.value

            # Energy Efficiency
            is_efficient = EEcheck.value
            efficiency_percent = EEreduction.value/100
            efficient_loads = EEloads.value

            # Flex
            is_flex = flexcheck.value
            flex_days = flexdays.value
            percent_flex_loads = flexloadpercent.value/100
            flex_reduction = flexpercent.value/100
            flex_only_noDH = True
            
            # Custom Loads
            custom = df_converter()
            
            optimal_flex = optimalcheck.value
            if flextoggle.value == 'Battery':
                maxkW = kW.value
                maxkWh = kWh.value
            else:
                maxkW = kW.value
                maxkWh = kWh.value

        substation = GUI_substation(region,
                                    resload_dict, 
                                    office_dict, 
                                    pv_dict,
                                    num_EV,
                                    num_parkingloc,
                                    mpg_mu,
                                    is_efficient, 
                                    efficiency_percent, 
                                    efficient_loads, 
                                    is_flex, 
                                    flex_days, 
                                    percent_flex_loads, 
                                    flex_reduction,
                                    flex_only_noDH,
                                    custom,
                                    optimal_flex,
                                    maxkW,
                                    maxkWh
                                   )
        
        
        #nameinfobox = widgets.Box([widgets.HTML(value='<h3> </h3>')],layout = {'margin':'15px'})

        consumptioninfobox = widgets.VBox([
                        widgets.HTML(value='<h4>General</h4>'),
                        widgets.HTML(value='Average hourly consumption [kWh]: {} (+/- {})'.format(substation.mu, substation.sigma)),
                        widgets.HTML(value='Maximum hourly consumption [kWh]: {}'.format(round(substation.dataframe['AggregatedLoad'].max(),3))),
                        widgets.HTML(value='Minimum hourly consumption [kWh]: {}'.format(round(substation.dataframe['AggregatedLoad'].min(),3))),
                        widgets.HTML(value='Percentiles and associated hourly consumption [kwh]:'),
                        widgets.VBox(children = [widgets.HTML('{} : {}'.format(key, round(value,3))) for key,value in substation.percentile_dict.items()], layout = {'margin':'0 0 0 30px'}),
                        widgets.HTML(value='Total consumption [kWh]: {}'.format(round(substation.dataframe['AggregatedLoad'].sum(),3))),
                        widgets.HTML(value='Based on data from {} to {}.'.format(substation.start,substation.end)),
                        widgets.HTML(value='The peak consumption occurs at: {}'.format(substation.find_max())),
                                      ],
                                     layout = {'margin':'10px'})

        countinfobox = widgets.VBox([
                                 widgets.HTML(value='<h4>Buildings</h4>'),
                                 widgets.HTML(value='Number of houses: {}'.format(substation.house_count)),
                                 widgets.HTML(value='Number of apartments: {}'.format(substation.apartment_count)),
                                 widgets.HTML(value='Number of offices: {}'.format(substation.office_count)),
                                 widgets.HTML(value='---- of which has district heating: {}'.format(substation.DH_count), layout={'margin':'0 0 0 60px'})],
                               layout = {'margin':'10px'})
        
        pvinfobox = widgets.VBox([widgets.HTML(value='<h4>PV</h4>'),
                                widgets.HTML(value='Number of PVs: {}'.format(substation.PV_count)),
                                widgets.HTML(value='With a total installed effect of {} kW.'.format((10.5 * pv_dict[69]) + (66 * pv_dict[400])+ (143 * pv_dict[868])+(pvsize.value * pvcustom.value))),],
                                layout = {'margin':'10px'})
        evinfobox = widgets.VBox([
                                widgets.HTML(value='<h4>EV</h4>'),
                                 widgets.HTML(value='Number of EVs: {}'.format(substation.EV_count)),
                                 widgets.HTML(value='Number of EV charging locations: {}'.format(num_parkingloc)),
                                 widgets.HTML(value='Average consumption of the EVs is {} kWh/km.'.format(mpg_mu)),
                                ],
                                layout = {'margin':'10px'})
        
        if substation.custom_count > 0:
            custominfobox = widgets.VBox([widgets.HTML(value='<h4>Custom Loads</h4>'),
                                          widgets.HTML(value='Number of custom loads : {}'.format(substation.custom_count)),
                                          ],
                                        layout = {'margin':'10px'})
        if is_efficient:
            eeinfobox = widgets.VBox([widgets.HTML(value='<h4>Energy Efficiency</h4>'),
                                    widgets.HTML(value='{} loads have taken energy efficient action which reduce their consumption with {} %.'.format(efficient_loads,int(efficiency_percent*100)))],
                                    layout = {'margin':'10px'})
        else:
            eeinfobox = widgets.VBox([widgets.HTML(value='<h4>Energy Efficiency</h4>'),
                                    widgets.HTML(value='No energy efficiency actions has been taken.')],
                                    layout = {'margin':'10px'})
            
        if is_flex:
            flexinfobox = widgets.VBox([widgets.HTML(value='<h4>Static Flex</h4>'),
                                      widgets.HTML(value='During the peak hours of {} days per year {} % of the loads were flexible.'.format(flex_days,int(percent_flex_loads*100))),
                                      widgets.HTML(value='The loads reduced their consumption with {} %.'.format(int(flex_reduction*100))),
                                      widgets.HTML(value='Hourly maximum consumption without flex [kWh]: {} '.format(round(substation.dataframe['NoFlex'].max(),3))),
                                      widgets.HTML(value='Hourly maximum consumption with flex [kWh]: {} '.format(round(substation.dataframe['AggregatedLoad'].max(),3)))
                                      ],layout = {'margin':'10px'})
        else:
            flexinfobox = widgets.VBox([widgets.HTML(value='<h4>Static Flex</h4>'),
                                       widgets.HTML(value='No demand side flexibility actions.'),
                                      ],layout = {'margin':'10px'})
            
        if optimal_flex:
            optimalflexinfobox = widgets.VBox([widgets.HTML(value='<h4>Optimal Flex</h4>'),
                                      widgets.HTML(value='Hourly maximum consumption without flex [kWh]: {} '.format(round(substation.dataframe['AggregatedLoad'].max(),3))),
                                      widgets.HTML(value='Hourly maximum consumption with optimal flex [kWh]: {} '.format(round(substation.dataframe['OptimalLoad'].max(),3))),
                                      widgets.HTML(value='Total consumption without flex [kWh]: {} '.format(round(sum(substation.dataframe['AggregatedLoad']),3))),
                                      widgets.HTML(value='Total consumption with optimal flex [kWh]: {} '.format(round(sum(substation.dataframe['OptimalLoad']),3))),      
                                              ],layout = {'margin':'10px'})  
        else:
            optimalflexinfobox = widgets.VBox([widgets.HTML(value='<h4>Optimal Flex</h4>'),
                                               widgets.HTML(value='No calculations of optimal load.'),
                                              ],layout = {'margin':'10px'})
        
        trendinfobox = widgets.VBox([pvinfobox,
                                     evinfobox,
                                     eeinfobox, 
                                     flexinfobox,
                                     optimalflexinfobox],)

        numinfobox = widgets.VBox([
                                   #nameinfobox, 
                                   consumptioninfobox, 
                                   countinfobox, 
                                   trendinfobox
                                  ])
        button_plot = widgets.Button(
                description='Plot',
                disabled=False,
                button_style='danger',
                tooltip='Click to Plot',
                )

        def plotter_clicked(b):
            plot()
    
        button_plot.on_click(plotter_clicked)
        
        ID_picker = widgets.Dropdown(
                    options = [i for i in range(1,substation.ID_count+1)],
                    value = 1, 
                    description = 'ID: ', 
                    )
        
        if optimal_flex:
            graph_type_options = ['Choose Graph Type','Load Duration Curve','Load Curve','Month Plot','Hour Plot','Weekday Plot','Single Load','Optimal Flex Graph']
        else:
            graph_type_options = ['Choose Graph Type','Load Duration Curve','Load Curve','Month Plot','Hour Plot','Weekday Plot','Single Load']
        
        graphout = widgets.Output()
        graph_type = widgets.Dropdown(
                options= graph_type_options,
                value='Choose Graph Type',
               # description='Graph Type:',
                disabled=False)
        
        start_picker = widgets.DatePicker(
                        description='Start date:',
                        disabled=False)
        end_picker = widgets.DatePicker(
                        description='End date:',
                        disabled=False)
        graphinfobox = widgets.VBox(children = [widgets.HBox([start_picker,
                                                end_picker]),
                                                widgets.HBox([graph_type,
                                                button_plot], layout = {'margin':'10px 0 0 20px'}), 
                                                graphout])
        def plot():
            graph = graph_type.value
            with graphout:
                graphout.clear_output()
                if (substation.dataframe is not None):
                    if start_picker.value:
                        if end_picker.value:
                            df = substation.dataframe[str(start_picker.value):str(end_picker.value)]
                        else: 
                            df = substation.dataframe[str(start_picker.value):]
                    else:
                        df = substation.dataframe
                    if graph == 'Load Duration Curve':
                        sorted_demand_list = df['AggregatedLoad'].sort_values(ascending=False).tolist()
                        list_len = len(sorted_demand_list) # Number of datapoints
                        x = np.linspace(1,list_len,list_len).tolist() # List of hours
                        fig = plt.plot(x,sorted_demand_list)
                        plt.title('Load duration curve')
                        plt.xlabel('Hours')
                        plt.ylabel('kWh')

                    elif graph == 'Load Curve':
                        fig1 = plt.plot(df['AggregatedLoad'])
                        plt.title('Aggregated load curve')
                        plt.ylabel('kWh')
                    elif graph == 'Month Plot':
                        ax = sns.boxplot(data=df, x='Month', y='AggregatedLoad')
                        ax.set_ylabel('kWh')
                        ax.set_title('Consumption by month')
                    elif graph == 'Hour Plot':
                        ax = sns.boxplot(data=df, x='Hour', y='AggregatedLoad')
                        ax.set_ylabel('kWh')
                        ax.set_title('Consumption by hour')
                    elif graph == 'Weekday Plot':
                        order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
                        ax1 = sns.boxplot(data=df, x='Weekday', y='AggregatedLoad', order=order)
                        ax1.set_ylabel('kWh')
                        ax1.set_xlabel('')
                        ax1.set_title('Consumption by day of the week')
                    elif graph == 'Single Load':
                        display(ID_picker)
                        substation.plot_single_load(ID = ID_picker.value, 
                                                    start = start_picker.value, 
                                                    end = end_picker.value)
                    elif graph == 'Optimal Flex Graph':
                        print('This graph shows the load before and after optimization. The load can be considered "optimally" distributed, given the specified constraints.')
                        plt.plot(df['AggregatedLoad'])
                        plt.plot(df['OptimalLoad'])
                        plt.legend(['Before','After'])
                        plt.title('Load curve before and after optimization of October to March')
                        plt.ylabel('kWh')
                    plt.show()
            

        tabout = widgets.Tab([numinfobox,graphinfobox])
        tabout.set_title(0, 'Info')
        tabout.set_title(1, 'Plots and Graphs')
        display(tabout)
        
    runbtn.on_click(lambda e: run_model(e,scenario = scenario_selection.value))
    runbtn2.on_click(lambda e: run_model(e))

    def flex_disable(event):
        if flexpercent.disabled:
            flexpercent.disabled = False
            flexdays.disabled = False
            flexloadpercent.disabled = False

        else:
            flexpercent.disabled = True
            flexdays.disabled = True
            flexloadpercent.disabled = True

    flexcheck.observe(lambda e: flex_disable(e))
    
    def optimal_disable(event):
        if kW.disabled:
            kW.disabled = False
            kWh.disabled = False
            flextoggle.disabled = False
        else:
            kW.disabled = True
            kWh.disabled = True
            flextoggle.disabled = True
    
    optimalcheck.observe(lambda e: optimal_disable(e))

    def change_flex_labels():
        if flextoggle.value == 'Battery':
            kWLabel.value = 'kW capacity'
            kWhLabel.value = 'kWh capxcity'
        else:
            kWLabel.value = 'kW of flexible agents'
            kWhLabel.value = 'kWh loads'
            
    flextoggle.observe(lambda e: change_flex_labels())

    def ee_disable(event):
        if EEloads.disabled:
            EEloads.disabled = False
            EEreduction.disabled = False
        else:
            EEloads.disabled = True
            EEreduction.disabled = True 

    EEcheck.observe(lambda e: ee_disable(e))

    def load_change(event):
        change = event['new'] - event['old']
        if EEloads.max + change > 0:
            EEloads.max += change

    housenew.observe(lambda e: load_change(e), names='value')
    houseold.observe(lambda e: load_change(e), names='value')
    housemix.observe(lambda e: load_change(e), names='value')
    apart.observe(lambda e: load_change(e), names='value')
    office800.observe(lambda e: load_change(e), names='value')
    office12402.observe(lambda e: load_change(e), names='value')
    office12667.observe(lambda e: load_change(e), names='value') 
    office28246.observe(lambda e: load_change(e), names='value') 
    office66799.observe(lambda e: load_change(e), names='value') 
    office73620.observe(lambda e: load_change(e), names='value')
    officecustom.observe(lambda e: load_change(e), names='value')
    
    def content_parser():
        if up.value == {}:
            return    
        else:
            typ, content = "", ""
            up_value = up.value
            for i in up_value.keys():
                typ = up_value[i]["metadata"]["type"]
                if typ == "text/csv":
                    content = up_value[i]["content"]
                    content_str = str(content, 'utf-8')
                    
                if content_str != "":
                    str_io = StringIO(content_str) 
                    return str_io
            
    def df_converter():
        content = content_parser()
        if content is not None:
            df = pd.read_csv(content, sep=',', index_col=0, parse_dates = True)    
            return df
        else:
            return pd.DataFrame()


    #----------------- Tabs and display --------------------------------------

    tab = widgets.Tab([scenariobox,create_scenario])
    tab.set_title(0, 'Scenario 1 - 4')
    tab.set_title(1, 'Custom Scenario')

    display(Markdown('# Model Simulation'), Markdown('Here you run the simulation. Either choose a scenario or build your own.'))
    display(tab)
```
